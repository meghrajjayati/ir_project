```python
import pandas as pd
file_path = "final_recipes/final_recipes.xlsx"
# Read the CSV file and handle the error by skipping the problematic row
# Read the Excel file into a pandas DataFrame
try:
    df = pd.read_excel(file_path)
    # Now 'df' contains the data from the Excel file
except pd.errors.ParserError as e:
    print(f"ParserError: {e}")
```

```python
columns_to_remove= ["Unnamed: 0"]
df = df.drop(columns=columns_to_remove, axis=1)
```

```python
pd.set_option('display.max_colwidth', None)  # Set to None for unlimited width
```

```python
import pandas as pd


df['ing'] = df['final_ing'].apply(lambda x: x.split(', '))
```

```python

# Function to find recipes with 3 or more common ingredients
def find_similar_recipes(recipe_id, threshold=1, top_n=5):
    ingredients_set = set(df.loc[recipe_id, 'ing'])
    similar_recipes = []

    for index, row in df.iterrows():
        if index != recipe_id:
            common_ingredients = set(row['ing']).intersection(ingredients_set)
            if len(common_ingredients) >= threshold:
                similar_recipes.append((index, common_ingredients))

    # Sort similar recipes by the number of common ingredients (descending order)
    similar_recipes.sort(key=lambda x: len(x[1]), reverse=True)

    # Take only the top N recipes
    top_n_recipes = similar_recipes[:top_n]

    return top_n_recipes

# Create a new DataFrame to store the desired structure
result_df = pd.DataFrame(columns=['Test_Query', 'Recipe_IDs'])

# Repeat the process for the first 30 recipe IDs
for recipe_id_to_check in range(0, 31):  # Assuming recipe IDs start from 1
    top_5_recipes = find_similar_recipes(recipe_id_to_check, threshold=3, top_n=5)
    test_query = df.loc[recipe_id_to_check, 'final_ing']
    recipe_ids = [recipe_id_to_check] + [recipe_id for recipe_id, common_ingredients in top_5_recipes]

    # Store information for the test query
    result_df.loc[recipe_id_to_check] = [test_query, recipe_ids]

# Print the result DataFrame

```
